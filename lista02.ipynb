# Exercício 01:

Em determinadas épocas do ano a venda de certos produtos sofre um aumento significativo. Um exemplo disso, são as vendas de sorvete que aumentam bastante no verão. Além do sorvete, outros itens como protetor solar e vestuário de banho podem ganhar maior atenção durante essa época do ano enquanto outros produtos podem não ser tão valorizados. Pensando nisso, crie uma função que, dado duas listas com as quantidades de produtos vendidos ao longo dos 12 meses do ano, crie um gráfico de barras ou linhas mostrando a evolução das vendas dos produtos ao longo dos meses em porcentagem. Ou seja, um gráfico relacionando a porcentagem de produtos vendidos naquele mês em relação ao ano como um todo para as vendas de sorvetes, óculos de sol e casacos.


```python
from matplotlib import pyplot as plt

#Exemplo:
ice_cream = [3000, 2600, 1400, 1500, 1200, 500, 300, 400, 700, 600, 800, 1900]
sunglasses = [1000, 800, 100, 70, 50, 190, 60, 50, 100, 120, 130, 900]
coats = [10, 20, 80, 120, 100, 500, 900, 780, 360, 100, 120, 20]
labels = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]

def bar_or_lines(list1, list2, list3):
    
    list4 = [x/sum(list1)*100 for x in list1]
    list5 = [x/sum(list2)*100 for x in list2]
    list6 = [x/sum(list3)*100 for x in list3]
        
    plt.plot(labels, list4, color='purple', marker='d', linestyle='--')
    plt.plot(labels, list5, color='coral', marker='o', linestyle=':')
    plt.plot(labels, list6, color='violet', marker='x', linestyle='-.')

    plt.xlabel("Meses do ano")
    plt.ylabel("% de itens vendidos")
    plt.title("Relação de itens vendidos por mês do ano")
    plt.show()

bar_or_lines(ice_cream, sunglasses, coats)
```

# Exercício 02:

Utilizando os mesmos dados do exercício anterior, crie uma função que, dado duas listas com as quantidades de produtos vendidos ao longo dos 12 meses do ano, crie um gráfico de dispersão mostrando a correlação entre as vendas dos dois produtos. Como exemplo, gere três gráficos, a partir das três listas que são dadas. Ou seja, um gráfico relacionando as vendas de sorvetes e óculos de sol, outro relacionando as vendas de sorvetes e casacos e, por fim, um relacionando as vendas de óculos de sol e casacos. Em seguida, discuta sobre os gráficos encontrados no sentido de encontrar uma relação entre as vendas.


__Dicas:__  
"_Correlação não é o mesmo que causalidade!_"


```python
from matplotlib import pyplot as plt

#Exemplo:
ice_cream = [3000, 2600, 1400, 1500, 1200, 500, 300, 400, 700, 600, 800, 1900]
sunglasses = [1000, 800, 100, 70, 50, 190, 60, 50, 100, 120, 130, 900]
coats = [10, 20, 80, 120, 100, 500, 900, 780, 360, 100, 120, 20]
labels = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]

def scatter(list1, list2):
    
    list3 = [x/sum(list1) for x in list1]
    list4 = [x/sum(list2) for x in list2]
    
    plt.scatter(list3, list4, c='red', marker='^')

    for label, product1, product2 in zip(labels, list3, list4):
        plt.annotate(label, xy=(product1, product2), xytext=(1, 5), textcoords='offset points')
        
    plt.title("Correlação entre vendas de dois produtos")
    plt.xlabel("% vendida do produto 1")
    plt.ylabel("% vendida do produto 2")
    plt.show()
        
scatter(ice_cream, sunglasses)
scatter(ice_cream, coats)
scatter(sunglasses, coats)
```

__Discuta sobre os resultados obtidos:__

Os três gráficos mostram correlações parecidas com relações de proporcionalidades, sendo a do primeiro gráfico diretamente proporcional e a dos dois últimos inversamente proporcional.

* **Gráfico 1:**

O primeiro gráfico mostra uma relação diretamente proporcional entre os dois produtos, porque pode-se notar que ambos valores estão normalmente próximos (estes valores ficariam próximos à reta x = y se os eixos fossem iguais) e aumentam ou diminuem em intensidades parecidas, dando a entender que os produtos são vendidos na mesma época (pico de venda nos meses de Dezembro, Janeiro e Fevereiro) e mantém uma porcentagem baixa e uniforme no restante do ano. Pode-se dizer que os meses do ano acompanham a mesma tendência da porcentagem de vendas, pois meses próximos estão localizados em regiões próximas e a variação nas vendas em um mês é propagada nos meses posteriores.  


* **Gráfico 2:**

A relação inversamente proporcional entre os dois produtos do gráfico 2 vem da ideia de que o aumento nas vendas de um produto vem acompanhado pela queda nas vendas do outro. O gráfico segue, assim como no gráfico 1, a ideia de que meses próximos refletem variações parecidas, mas o meses já não estão divididos em agrupamentos, dando e entender que as vendas são mais uniformes e sutis, crescendo em épocas específicas e distintas para cada um dos produtos. 

* **Gráfico 3:**

Para o terceiro gráfico, a relação inversamente proporcional fica ainda mais explicitada que no gráfico 2 pois existem agrupamentos de meses (assim como no gráfico 1), onde há um grupo num pico de venda para cada produto, isto é, cada produto tem um ponto onde suas vendas têm porcentagem alta e a do outro é muito mais baixa, se comparado com o restante dos meses no gráfico onde os valores são mais próximos, menores e mais uniformes.


# Exercício 03:


Crie uma matriz `n x m` com valores aleatórios positivos e negativos e, em seguida, gere um mapa de calor (_Heatmap_) com uma escala de cores gerada de forma que a cor branca represente o 0.


```python
import numpy as np
import matplotlib

#Exemplo:
A = [[6, -3], [-4, 8]]
B = [[1,2,3],[4,5,6],[7,8,9]]

# matriz aleatória (tamanho mínimo 2 x 2, tamanho máximo 10 x 10)
# valores internos gerados de -10 a 10
n = np.random.randint(2, 10)
m = np.random.randint(2, 10)
randomMatrix = np.random.randint(-10, 10, size=(n,m))

# como não sabia se era pra criar a matriz e passar para função ou fazer uma função que aceitasse qualquer matriz, 
# fiz ambas

def heatmap(matrix):
    maxM = np.amax(matrix)
    minM = - np.amax(matrix)
    
    plt.imshow(matrix, cmap=matplotlib.cm.Spectral, vmin=minM, vmax=maxM)
    plt.colorbar()
    plt.show()

heatmap(A)
heatmap(randomMatrix)
```
